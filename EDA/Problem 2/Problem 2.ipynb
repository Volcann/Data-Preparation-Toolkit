```python
# Import the nessasary modules
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

<h2 style="color: lightblue;">üõë Load the Dataset</h2>

```python
# Load the dataset
data = pd.read_csv('Iris.csv')
data
```

<h2 style="color: lightblue;">üõë Univariate Analysis</h2>

```python
# Plot the distribution of 'Age'
sns.histplot(data['SepalLengthCm'], kde=True)
plt.title('Age Distribution')
plt.show()

# Plot the distribution of 'Fare'
sns.histplot(data['SepalWidthCm'], kde=True)
plt.title('Fare Distribution')
plt.show()
```

```python
# Plot the distribution of 'Age'
sns.histplot(data['PetalLengthCm'], kde=True)
plt.title('Age Distribution')
plt.show()

# Plot the distribution of 'Fare'
sns.histplot(data['PetalWidthCm'], kde=True)
plt.title('Fare Distribution')
plt.show()
```

```python
# Distribution of numerical features
data.hist(bins=20, figsize=(10, 8))
plt.suptitle('Histogram of Numerical Features')
plt.show()

# Boxplot of numerical features
plt.figure(figsize=(10, 6))
sns.boxplot(data=data)
plt.title('Boxplot of Numerical Features')
plt.show()
```

<h2 style="color: lightblue;">üõë Bivariate Analysis</h2>

```python
# Ensure correct column name for hue
sns.pairplot(data, hue='Species', markers=["o", "s", "D"])
plt.suptitle('Pairplot of Iris Dataset')
plt.show()
```

<h2 style="color: lightblue;">üõë Correlation Analysis</h2>

```python
# Exclude non-numeric columns
numeric_data = data.select_dtypes(include=['number'])

# Compute the correlation matrix
corr_matrix = numeric_data.corr()

# Print the correlation matrix
print("\nCorrelation matrix:")
print(corr_matrix)

# Heatmap of the correlation matrix
plt.figure(figsize=(8, 6))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title('Heatmap of Correlation Matrix')
plt.show()
```
